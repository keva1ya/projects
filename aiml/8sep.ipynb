```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.arange(-60, 1, 0.1)
y1 = np.exp(x1)


x2 = np.arange(-4 * np.pi, 4 * np.pi + np.pi/15, np.pi/15)
y2 = np.sin(x2)

plt.figure(figsize=(10,6))
plt.plot(x1, y1, label='y = exp(x)', color='blue')
plt.plot(x2, y2, label='y = sin(x)', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = exp(x) and y = sin(x)')
plt.legend()
plt.grid(True)
plt.axhline(0,color='black',linewidth=1)
plt.show()
```
